```python
bcd = sklearn.datasets.load_breast_cancer()
```

```python
print(bcd)
```

```python
data_frame = pd.DataFrame(bcd.data,columns=bcd.feature_names)
```

```python
data_frame['label'] = bcd.target
```

```python
# Now  here we  are seperating the features and target
X = data_frame.drop(columns='label',axis =1)
Y = data_frame['label']
```

```python
# Now we aare giving seperate the  features and  target to the model

# spliting data into train data  and  split data
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
print(X.shape ,  X_test.shape, X_train.shape)
```

```python
# Now we are standardizing the input means only train aand test of X
from sklearn.preprocessing import StandardScaler
```

```python
scaler  = StandardScaler()

X_train_std= scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)
```

Building the neural  network


```python
import tensorflow as tf
tf.random.set_seed(4)
from tensorflow import keras
```

```python
# setting up layers for my model there are difrrent layers input hidden and one output layer
model = keras.Sequential([

                          keras.layers.Flatten(input_shape=(30,)),
                          keras.layers.Dense(20,activation='relu'),

                          keras.layers.Dense(2,activation='sigmoid')
])
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
history = model.fit(X_train, Y_train, validation_split=0.1, epochs=10)
```

```python
history_std = model.fit(X_train_std, Y_train, validation_split=0.1, epochs=10)
```

```python
# Now you can see  the  accuracy is more in the scaler one now we are visualizee them
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'upper right')
```

```python
loss, accuracy = model.evaluate(X_test_std, Y_test)
print(accuracy)
```
